## Logarithm versus exponential

$$ \ln(x) = y \iff x = e^y $$

$$ e^x = y \iff x = \ln(y) $$


```python
x_exp = np.linspace(-6, 1.8, 100)
x_log = np.linspace(0.01, 6, 100)
y_exp = np.exp(x_exp)
y_log = np.log(x_log)
x_linear = np.linspace(-6, 6, 100)  # Match the range of other functions
y_linear = x_linear  # f(x) = x is just the identity function

plt.figure(figsize=(8, 8))
plt.plot(x_exp, y_exp, label='exp(x)', color='blue')
plt.plot(x_log, y_log, label='ln(x)', color='red')
plt.plot(x_linear, y_linear, label='x', color='green')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=1)
plt.axvline(x=0, color='k', linestyle='-', alpha=1)
plt.legend()
plt.title('Natural Logarithm vs Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')  # Make the scale equal on both axes
plt.show()
```

The logarithm function is the inverse of the exponential function. They are symmetric with respect to the line y = x.
$$ \ln(e^x) = x $$
